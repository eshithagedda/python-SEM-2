```python
import matplotlib.pyplot as plt
import numpy as np

mylabels=["maths","english","chemistry","beee","python"]
mycolors=["black","pink","purple","c","darkgreen"]
y = np.array([20, 19, 18, 20,20])
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels=mylabels, startangle=180,colors=mycolors)
plt.legend(title="MARKS: ")
plt.show() 
```
